```python
class Employee():
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def get_info(self):
        return f'Имя: {self.name}, ID: {self.id}'

class Manager(Employee):
    def __init__(self, name, id, department):
        super().__init__(name, id)
        self.department = department

    def manage_project(self):
        print(f'{self.name} управляет проектом в отделе {self.department}')

class Technician(Employee):
    def __init__(self, name, id, specialization):
        super().__init__(name, id)
        self.specialization = specialization

    def perform_maintenance(self):
        print(f'{self.name} выполняет техническое обслуживание')

class TechManager(Manager, Technician):
    def __init__(self, name, id, specialization, department):
        self.name = name
        self.id = id
        self.specialization = specialization
        self.department = department
        self.employees = []

    def add_employee(self, employee):
        self.employees.append(employee)

    def get_team_info(self):
        team_info = [f'{employee.get_info()}\n' for employee in self.employees]
        return ''.join(team_info)



john = Employee(name='Джон', id=1001)
mary = Manager(name='Мэри', id=2001, department='Маркетинг')
bob = Technician(name='Боб', id=3001, specialization='Электроника')
alice = TechManager(name='Элис', id=4001, department='Техническая поддержка', specialization='Программирование')

print(john.get_info())

print(mary.get_info())
mary.manage_project()

print(bob.get_info())
bob.perform_maintenance()

print(alice.get_info())
alice.manage_project()
alice.perform_maintenance()

alice.add_employee(john)
alice.add_employee(mary)
alice.add_employee(bob)
print(alice.get_team_info())
```
